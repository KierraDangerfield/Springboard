# Storytelling with Dogs of Zurich

```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#get the data and read it 
dogData = pd.read_csv('./20170308hundehalter.csv')
```

```python
#lets view
dogData.head(10)
```

ALTER -> Age

GESCHLECHT -> Gender

STADTKREIS -> City Quarter or District

RASSE1 -> Dog's Primary Breed

RASSE2 -> Dog's Secondary Breed

GEBURTSJAHR_HUND -> Dog's Year of Birth

GESCHLECHT_HUND -> Dog's Gender

HUNDEFARBE -> Dog's Colo

```python
#let's rename the columns
dogData.rename(columns={'ALTER': "Owner's Age",
                       'GESCHLECHT': "Owner's Gender",
                       'STADTKREIS': 'City Quarter or District',
                       'RASSE1': "Dog's Primary Breed",
                       'RASSE2': "Dog's Secondary Breed",
                       'GEBURTSJAHR_HUND': "Dog's Birth Year",
                       'GESCHLECHT_HUND': "Dog's Gender",
                       'HUNDEFARBE': "Dog's Colo"}, inplace=True)

dogData
```

```python
#get dog age 
dogData['dog_age'] = 2017 - dogData["Dog's Birth Year"]
dogData
```

```python
#Owners' age vs Dogs' Age
fig, ax = plt.subplots(figsize=(15,10))
ax = sns.barplot(y="dog_age", x=dogData["Owner's Age"].sort_values(ascending=True), hue= "Owner's Gender", data=dogData).set(title="Owners' Age vs Dogs' Age")
```

```python
#Top breeds
topDog = dogData["Dog's Primary Breed"].value_counts()
topDog.head(20)
```

```python
fig, ax = plt.subplots(figsize=(15,10))

topDog[0:20].plot(kind='bar', title="Top 20 dog breeds")
```

# Top Dogs Based on Age Group

### Top breeds for 11-20 Age Group

```python
#Top breeds for 11-20
elevon_20 = dogData[dogData["Owner's Age"] == "11-20"]

fig, ax = plt.subplots(figsize=(15,10))
ax = sns.countplot(x="Dog's Primary Breed", hue= "Owner's Gender", data=elevon_20,
                  order=elevon_20["Dog's Primary Breed"].value_counts().iloc[:15].index).set(
    title="Owners' Age: 11-20 vs Dogs' Breed")

plt.xticks(rotation = 45)
plt.show()
```
