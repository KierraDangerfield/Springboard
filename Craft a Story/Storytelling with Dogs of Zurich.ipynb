# Storytelling with Dogs of Zurich

```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#get the data and read it 
dogData = pd.read_csv('./20170308hundehalter.csv')
```

```python
#lets view
dogData.head(10)
```

ALTER -> Age

GESCHLECHT -> Gender

STADTKREIS -> City Quarter or District

RASSE1 -> Dog's Primary Breed

RASSE2 -> Dog's Secondary Breed

GEBURTSJAHR_HUND -> Dog's Year of Birth

GESCHLECHT_HUND -> Dog's Gender

HUNDEFARBE -> Dog's Colo

```python
#let's rename the columns
dogData.rename(columns={'ALTER': "Owner's Age",
                       'GESCHLECHT': "Owner's Gender",
                       'STADTKREIS': 'City Quarter or District',
                       'RASSE1': "Dog's Primary Breed",
                       'RASSE2': "Dog's Secondary Breed",
                       'GEBURTSJAHR_HUND': "Dog's Birth Year",
                       'GESCHLECHT_HUND': "Dog's Gender",
                       'HUNDEFARBE': "Dog's Colo"}, inplace=True)

dogData
```

```python
#get dog age 
dogData['dog_age'] = 2017 - dogData["Dog's Birth Year"]
dogData
```

```python
#Owners' age vs Dogs' Age
fig, ax = plt.subplots(figsize=(15,10))
ax = sns.barplot(y="dog_age", x=dogData["Owner's Age"].sort_values(ascending=True), hue= "Owner's Gender", data=dogData).set(title="Owners' Age vs Dogs' Age")
```

```python
#Top breeds
topDog = dogData["Dog's Primary Breed"].value_counts()
topDog.head(20)
```

```python
fig, ax = plt.subplots(figsize=(15,10))

topDog[0:20].plot(kind='bar', title="Top 20 dog breeds")
```

# Top Dogs Based on Age Group

## Top breeds based on 11-20, 21-30, 31-40 age group

```python
#Top breeds based on 31-40 age group
elevon_20 = dogData[dogData["Owner's Age"] == "11-20"]
twentyone_30 = dogData[dogData["Owner's Age"] == "21-30"]
thirtyone_40 = dogData[dogData["Owner's Age"] == "31-40"]


fig, (ax1, ax2, ax3) = plt.subplots(ncols=1, nrows=3, figsize=(15,25))
fig.suptitle("Age Groups vs Dog Breeds")


sns.countplot(x="Dog's Primary Breed", data=elevon_20, hue="Owner's Gender", 
              order=elevon_20["Dog's Primary Breed"].value_counts().iloc[:15].index, 
              ax=ax1).set(title="Owners' Age: 11-20 vs Dogs' Breed")

sns.countplot(x="Dog's Primary Breed", data=twentyone_30, hue="Owner's Gender", 
              order=twentyone_30["Dog's Primary Breed"].value_counts().iloc[:15].index, 
              ax=ax2).set(title="Owners' Age:  21-30 vs Dogs' Breed")

sns.countplot(x="Dog's Primary Breed", data=thirtyone_40, hue="Owner's Gender", 
              order=thirtyone_40["Dog's Primary Breed"].value_counts().iloc[:15].index,
              ax=ax3).set(title="Owners' Age: 31-40 vs Dogs' Breed")

plt.rcParams['font.size'] = '16'

ax1.tick_params(axis='x', labelrotation=45)
ax2.tick_params(axis='x', labelrotation=45)
ax3.tick_params(axis='x', labelrotation=45)

fig.tight_layout()
plt.show()
```

## Top breeds based on 41-50,  51-60, 61-70  age group

```python
#Top breeds based on 61-70 age group
fortyone_50 = dogData[dogData["Owner's Age"] == "41-50"]
fiftyone_60 = dogData[dogData["Owner's Age"] == "51-60"]
sixtyone_70 = dogData[dogData["Owner's Age"] == "61-70"]


fig, (ax1, ax2, ax3) = plt.subplots(ncols=1, nrows=3, figsize=(15,25))
fig.suptitle("Age Groups vs Dog Breeds")


sns.countplot(x="Dog's Primary Breed", data=fortyone_50, hue="Owner's Gender", 
              order=fortyone_50["Dog's Primary Breed"].value_counts().iloc[:15].index, 
              ax=ax1).set(title="Owners' Age: 41-50 vs Dogs' Breed")

sns.countplot(x="Dog's Primary Breed", data=fiftyone_60, hue="Owner's Gender", 
              order=fiftyone_60["Dog's Primary Breed"].value_counts().iloc[:15].index, 
              ax=ax2).set(title="Owners' Age:  51-60 vs Dogs' Breed")

sns.countplot(x="Dog's Primary Breed", data=sixtyone_70, hue="Owner's Gender", 
              order=sixtyone_70["Dog's Primary Breed"].value_counts().iloc[:15].index,
              ax=ax3).set(title="Owners' Age: 61-70 vs Dogs' Breed")

plt.rcParams['font.size'] = '16'

ax1.tick_params(axis='x', labelrotation=45)
ax2.tick_params(axis='x', labelrotation=45)
ax3.tick_params(axis='x', labelrotation=45)

fig.tight_layout()
plt.show()
```

```python
#Top breeds vs cities

eightyone_90 = dogData[dogData["Owner's Age"] == "81-90"]
ninetyone_100 = dogData[dogData["Owner's Age"] == "91-100"]
seventyone_80 = dogData[dogData["Owner's Age"] == "71-80"]


fig, (ax1, ax2, ax3) = plt.subplots(ncols=1, nrows=3, figsize=(15,25))
fig.suptitle("Age Groups vs Dog Breeds")


sns.countplot(x="Dog's Primary Breed", data=seventyone_80, hue="Owner's Gender", 
              order=seventyone_80["Dog's Primary Breed"].value_counts().iloc[:15].index, 
              ax=ax1).set(title="Owners' Age: 71-80 vs Dogs' Breed")
#ax[0].xticks(rotation = 45)

sns.countplot(x="Dog's Primary Breed", data=eightyone_90, hue="Owner's Gender", 
              order=eightyone_90["Dog's Primary Breed"].value_counts().iloc[:15].index, 
              ax=ax2).set(title="Owners' Age:  81-90 vs Dogs' Breed")

sns.countplot(x="Dog's Primary Breed", data=ninetyone_100, hue="Owner's Gender", 
              order=ninetyone_100["Dog's Primary Breed"].value_counts().iloc[:15].index,
              ax=ax3).set(title="Owners' Age: 91-100 vs Dogs' Breed")

plt.rcParams['font.size'] = '16'

ax1.tick_params(axis='x', labelrotation=45)
ax2.tick_params(axis='x', labelrotation=45)
ax3.tick_params(axis='x', labelrotation=45)

fig.tight_layout()

plt.show()
```
